```python
%%writefile app.py
import streamlit as st
import tensorflow as tf
from PIL import Image, ImageOps
import numpy as np
import os
import requests
from io import BytesIO

MODEL_PATH = '/content/drive/MyDrive/teeth_classification/outputs/models/week2/mobilenetv2_best.keras'
IMG_SIZE = (224, 224)
CLASS_NAMES = ['CaS', 'CoS', 'Gum', 'MC', 'OC', 'OLP', 'OT']

st.set_page_config(page_title="Teeth Classifier", page_icon="ü¶∑")
st.title("ü¶∑ Teeth Classification AI")

@st.cache_resource
def load_model():
    return tf.keras.models.load_model(MODEL_PATH)

# Upload or demo image
image = None
file = st.file_uploader("Upload Image", type=["jpg", "jpeg", "png"])
if file:
    image = Image.open(file)
    st.image(image, caption='Uploaded Image', width="content")

if image is not None:
    model = load_model()
    img = ImageOps.fit(image.convert("RGB"), IMG_SIZE, Image.Resampling.LANCZOS)
    img_array = np.expand_dims(np.array(img), axis=0)
    preds = model.predict(img_array)
    cls = CLASS_NAMES[np.argmax(preds[0])]
    conf = np.max(preds[0])
    st.success(f"**Prediction:** {cls}")
    st.info(f"**Confidence:** {conf*100:.2f}%")
    st.bar_chart(dict(zip(CLASS_NAMES, preds[0])))
```
